```python
def generate_n_targets(df, noise=1, p1=0.5, p2=0.5, n_targets=10):
    coeffs=[]
    n_data = df.shape[0]
    
    y = np.zeros((n_data,n_targets))
    
    ### standardize features dataframe
    scaler = preprocessing.StandardScaler().fit(df)
    x = scaler.transform(df)
    df = pd.DataFrame(x,columns=df.columns)
    
    for i in range(n_targets):
        coeffs.append(np.random.uniform(low=-1,high=1,size=df.shape[1]))

        epsilon = np.random.normal(0, noise, size=(df.shape[0],1))
        y[:,i] = np.dot(x,coeffs[-1]).reshape(-1,) + epsilon.reshape(-1,)
        
    return y,coeffs


```

### Third Example: correlated variances

```python
def generate_n_targets(df, noise=1, p1=0.5, p2=0.5, n_targets=10):
    coeffs=[]
    n_data = df.shape[0]
    
    y = np.zeros((n_data,n_targets))
    
    common_noise_1 = np.random.normal(0, noise*p1, size=(df.shape[0],1))
    common_noise_2 = np.random.normal(0, noise*p1, size=(df.shape[0],1))
    
    ### standardize features dataframe
    scaler = preprocessing.StandardScaler().fit(df)
    x = scaler.transform(df)
    df = pd.DataFrame(x,columns=df.columns)
    
    for i in range(n_targets):
        if i%2==0:
            coeffs.append(np.random.uniform(low=0.5,high=1,size=df.shape[1]))
            epsilon = common_noise_1 + np.random.normal(0, noise*p2, size=(df.shape[0],1))
        else:
            coeffs.append(np.random.uniform(low=-1,high=-0.5,size=df.shape[1]))
            epsilon = common_noise_2 + np.random.normal(0, noise*p2, size=(df.shape[0],1))
        
        y[:,i] = np.dot(x,coeffs[-1]).reshape(-1,) + epsilon.reshape(-1,)
        
    return y,coeffs
```

```python
y,coeffs = generate_n_targets(df, noise=10, p1=0.5, p2=0.5, n_targets=10)
y[:,-1] = y[:,-2] + np.random.normal(0, 10, size=(df.shape[0],1)).reshape(-1,)
y_df = pd.DataFrame(y)
#y_df = (y_df-y_df.mean(axis=0))

myFun = NonLinCTA(df=df, targets_df=y_df, eps=0, n_val=-1, neigh=0)
myFun.compute_target_clusters()
```
